````python
import pandas as pd
import numpy as np

# Define the years for analysis (2010 - 2030)
years = list(range(2010, 2030))

# Generate synthetic market data for Finland & Norway
np.random.seed(42)
data = []

for year in years:
    for country in ["Finland", "Norway"]:
        market_size = np.random.randint(50, 150)  # Tech industry market size in billion $
        growth_rate = np.random.uniform(1.5, 5.5)  # % growth rate
        tech_investment = np.random.randint(10, 50)  # Investments in billion $
        startup_success_rate = np.random.uniform(20, 70)  # % of startups surviving
        innovation_index = np.random.uniform(60, 90)  # Innovation score out of 100

        data.append([year, country, market_size, growth_rate, tech_investment, startup_success_rate, innovation_index])

# Create DataFrame
df_market = pd.DataFrame(data, columns=["Year", "Country", "Market Size ($B)", "Growth Rate (%)",
                                        "Tech Investment ($B)", "Startup Success Rate (%)", "Innovation Index"])

# Save data to CSV
df_market.to_csv("tech_market_data.csv", index=False)

# Preview the data
print(df_market.head())
````

````python
# List of tech companies (simulated)
tech_companies = ["Company A", "Company B", "Company C", "Company D", "Company E"]

data_competitors = []

for company in tech_companies:
    for country in ["Finland", "Norway"]:
        revenue = np.random.randint(1, 20)  # Revenue in billion $
        market_share = np.random.uniform(5, 30)  # Market share percentage
        r_d_spending = np.random.randint(100, 500)  # R&D spending in million $
        customer_satisfaction = np.random.uniform(50, 90)  # Satisfaction score

        data_competitors.append([company, country, revenue, market_share, r_d_spending, customer_satisfaction])

# Create DataFrame
df_competitors = pd.DataFrame(data_competitors, columns=["Company", "Country", "Revenue ($B)",
                                                          "Market Share (%)", "R&D Spending ($M)", "Customer Satisfaction (%)"])

# Save data to CSV
df_competitors.to_csv("tech_competitor_data.csv", index=False)

# Preview the data
print(df_competitors.head())
````

````python
# Generate Business Performance Data
data_business = []

for year in years:
    for country in ["Finland", "Norway"]:
        total_sales = np.random.randint(1000, 5000)  # Total number of products/services sold
        revenue = np.random.randint(10, 50)  # Revenue in billion $
        profit_margin = np.random.uniform(10, 30)  # Profit margin in %
        customer_growth = np.random.uniform(1.5, 10)  # Annual customer growth %

        data_business.append([year, country, total_sales, revenue, profit_margin, customer_growth])

# Create DataFrame
df_business = pd.DataFrame(data_business, columns=["Year", "Country", "Total Sales", "Revenue ($B)",
                                                   "Profit Margin (%)", "Customer Growth (%)"])

# Save data to CSV
df_business.to_csv("business_performance_data.csv", index=False)

# Preview the data
print(df_business.head())
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA

# Change this:
#from fbprophet import Prophet

# To this:
from prophet import Prophet

# Filter data for forecasting
df_forecast = df_final.groupby(["Year", "Country"])["Market Size ($B)"].sum().reset_index()

# Rename columns for Prophet
df_forecast.rename(columns={"Year": "ds", "Market Size ($B)": "y"}, inplace=True)

# Plot the historical data
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_forecast, x="ds", y="y", hue="Country", marker="o")
plt.title("Historical Market Size Growth")
plt.xlabel("Year")
plt.ylabel("Market Size ($B)")
plt.show()
````

````python
from statsmodels.tsa.arima.model import ARIMA

# Select Finland's data for ARIMA
df_finland = df_forecast[df_forecast["Country"] == "Finland"].set_index("ds")

# Train ARIMA Model
model = ARIMA(df_finland["y"], order=(2,1,2))
model_fit = model.fit()

# Forecast the next 5 years
forecast_years = 5
forecast = model_fit.forecast(steps=forecast_years)

# Plot Forecast
plt.figure(figsize=(12,6))
plt.plot(df_finland.index, df_finland["y"], label="Historical")
plt.plot(pd.date_range(start=df_finland.index[-1], periods=forecast_years+1, freq="YE")[1:], forecast, label="Forecast", linestyle="dashed")
plt.title("ARIMA Forecast for Finland's Market Size")
plt.xlabel("Year")
plt.ylabel("Market Size ($B)")
plt.legend()
plt.show()
````

````python
# Train Prophet Model
prophet_model = Prophet()
prophet_model.fit(df_forecast)

# Create Future Dataframe for Prediction
future = prophet_model.make_future_dataframe(periods=5, freq="YE")
forecast = prophet_model.predict(future)

# Plot Forecast
prophet_model.plot(forecast)
plt.title("Prophet Forecast for Nordic Tech Market")
plt.show()
````

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate Hypothetical GDP Data for 20 Years
np.random.seed(42)
years = list(range(2003, 2023))

# Ensure all lists have the same length (40 in this case)
data = {
    "Year": years * 2,  # Duplicate years for both countries
    "Country": ["Finland"] * 20 + ["Norway"] * 20,
    "GDP ($B)": np.random.uniform(200, 400, 20).tolist() + np.random.uniform(300, 600, 20).tolist(),
    "Inflation (%)": np.random.uniform(0.5, 5.0, 40),
    "Unemployment (%)": np.random.uniform(3.0, 10.0, 40),
    "Interest Rate (%)": np.random.uniform(0.5, 5.0, 40),
    "Population (M)": np.random.uniform(5.5, 6.0, 20).tolist() + np.random.uniform(5.2, 5.8, 20).tolist(),
    "Market Size ($B)": np.random.uniform(50, 150, 40),
}

df_gdp = pd.DataFrame(data)

# Visualizing GDP Growth
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_gdp, x="Year", y="GDP ($B)", hue="Country", marker="o")
plt.title("GDP Growth in Finland & Norway (2003-2023)")
plt.xlabel("Year")
plt.ylabel("GDP ($B)")
plt.show()
````

````python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

# Select Features & Target
features = ["Inflation (%)", "Unemployment (%)", "Interest Rate (%)", "Population (M)", "Market Size ($B)"]
target = "GDP ($B)"

# Split Data into Training & Testing
df_finland = df_gdp[df_gdp["Country"] == "Finland"]
X_train, X_test, y_train, y_test = train_test_split(df_finland[features], df_finland[target], test_size=0.2, random_state=42)

# Train Linear Regression Model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
y_pred_lr = lr_model.predict(X_test)

# Train Random Forest Model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)

# Compare Performance
print("Linear Regression MAE:", mean_absolute_error(y_test, y_pred_lr))
print("Random Forest MAE:", mean_absolute_error(y_test, y_pred_rf))

# Show R² Score
print("Linear Regression R²:", r2_score(y_test, y_pred_lr))
print("Random Forest R²:", r2_score(y_test, y_pred_rf))
````

````python
import dash
from dash import dcc, html, Input, Output
import plotly.express as px
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor

# For demonstration, we'll assume you have already loaded your historical GDP and future projection data.
# For example, you might have already loaded:
# df_gdp = pd.read_csv("df_gdp.csv")
# df_future = pd.read_csv("df_future.csv")
# In this example, I'll simulate some data:

# Simulated historical GDP data for Finland and Norway
years_hist = list(range(2003, 2023))
data_hist = []
for country in ["Finland", "Norway"]:
    for year in years_hist:
        data_hist.append({
            "Year": year,
            "GDP ($B)": np.random.uniform(200, 400),
            "Country": country,
            "Inflation (%)": np.random.uniform(0.5, 5.0),
            "Unemployment (%)": np.random.uniform(3.0, 10.0),
            "Interest Rate (%)": np.random.uniform(0.5, 5.0),
            "Market Size ($B)": np.random.uniform(50, 150)
        })
df_gdp = pd.DataFrame(data_hist)

# Simulated future GDP projections (we'll generate dummy data for 2024-2028)
future_years = list(range(2024, 2029))
data_future = []
for country in ["Finland", "Norway"]:
    for year in future_years:
        data_future.append({
            "Year": year,
            "GDP ($B)": np.nan,  # We'll predict these values later
            "Country": country
        })
df_future = pd.DataFrame(data_future)

# Train a RandomForestRegressor on historical data for demonstration purposes.
# We'll use a simple model based on a few features.
features = ["Inflation (%)", "Unemployment (%)", "Interest Rate (%)", "Market Size ($B)"]
target = "GDP ($B)"
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
# For training, we'll filter for one country (e.g., Finland) for simplicity.
df_train = df_gdp[df_gdp["Country"] == "Finland"].dropna()
rf_model.fit(df_train[features], df_train[target])

# Concatenate historical and future data for the initial figure using pd.concat
combined_df = pd.concat([df_gdp, df_future], ignore_index=True)
fig_gdp = px.line(
    combined_df,
    x="Year",
    y="GDP ($B)",
    color="Country",
    markers=True,
    title="GDP Growth & Forecast for Finland & Norway"
)

# Initialize Dash App
app = dash.Dash(__name__)

app.layout = html.Div(children=[
    html.H1("Nordic Economic Dashboard"),

    # Country Selection Dropdown
    dcc.Dropdown(
        id="country-dropdown",
        options=[
            {"label": "Finland", "value": "Finland"},
            {"label": "Norway", "value": "Norway"}
        ],
        value="Finland",
        clearable=False
    ),

    # Sliders for Economic Indicators
    html.Label("Inflation Rate (%)"),
    dcc.Slider(id="inflation-slider", min=0.5, max=5.0, step=0.1, value=2.0, marks={i: f"{i}%" for i in range(1, 6)}),
    html.Label("Unemployment Rate (%)"),
    dcc.Slider(id="unemployment-slider", min=3.0, max=10.0, step=0.5, value=5.0, marks={i: f"{i}%" for i in range(3, 11)}),
    html.Label("Interest Rate (%)"),
    dcc.Slider(id="interest-slider", min=0.5, max=5.0, step=0.1, value=2.5, marks={i: f"{i}%" for i in range(1, 6)}),
    html.Label("Market Size ($B)"),
    dcc.Slider(id="market-slider", min=50, max=200, step=10, value=100, marks={i: str(i) for i in range(50, 201, 25)}),

    # GDP Graph
    dcc.Graph(id="gdp-graph", figure=fig_gdp)
])

# Callback to update the GDP graph based on user inputs
@app.callback(
    Output("gdp-graph", "figure"),
    [Input("country-dropdown", "value"),
     Input("inflation-slider", "value"),
     Input("unemployment-slider", "value"),
     Input("interest-slider", "value"),
     Input("market-slider", "value")]
)
def update_graph(selected_country, inflation, unemployment, interest_rate, market_size):
    # Filter historical data for the selected country
    df_filtered = df_gdp[df_gdp["Country"] == selected_country].copy()

    # Prepare future data based on user inputs
    future_data = pd.DataFrame({
        "Inflation (%)": [inflation] * len(future_years),
        "Unemployment (%)": [unemployment] * len(future_years),
        "Interest Rate (%)": [interest_rate] * len(future_years),
        "Market Size ($B)": [market_size] * len(future_years)
    })

    # Predict future GDP using the trained model
    future_gdp = rf_model.predict(future_data)
    df_future_filtered = pd.DataFrame({
        "Year": future_years,
        "GDP ($B)": future_gdp,
        "Country": [selected_country] * len(future_years)
    })

    # Concatenate historical and predicted future data using pd.concat
    combined_df_new = pd.concat([df_filtered, df_future_filtered], ignore_index=True)

    # Create updated figure
    fig = px.line(
        combined_df_new,
        x="Year",
        y="GDP ($B)",
        color="Country",
        markers=True,
        title=f"GDP Growth & Forecast for {selected_country}"
    )
    return fig

if __name__ == "__main__":
    app.run(debug=True)
````
